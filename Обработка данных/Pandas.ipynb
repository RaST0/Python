# Группировка данных

```python
group1 = df[df.Nationality == 'Spain'].Age
group2 = df[df.Nationality == 'Argentina'].Age
group3 = df[df.Nationality == 'Brazil'].Age

df_new = pd.DataFrame([group1.mean(),group2.mean(),group3.mean()], columns=['Age'], index = ['Spain','Argentina','Brazil'])
df_new
```

```python
df_new = pd.DataFrame({
    'mean' : [group1.mean(),group2.mean(),group3.mean()],
    'min' : [group1.min(),group2.min(),group3.min()],
    'max' : [group1.max(),group2.max(),group3.max()],
},index = ['Spain','Argentina','Brazil'])
df_new
```

```python
# Показать среднюю ЗП футболистов в любых 4-х клубах, клуб = AS Monaco
df[df.Club == 'AS Monaco'].Wage.mean()
```

```python
# Показать среднюю ЗП футболистов в любых 4-х клубах, клуб = Juventus
df[df.Club == 'Juventus'].Wage.mean()
```

```python
# Показать среднюю ЗП футболистов в любых 4-х клубах, клуб = Manchester City
df[df.Club == 'Manchester City'].Wage.mean()
```

```python
# Показать среднюю ЗП футболистов в любых 4-х клубах
group1 = df[df.Club == 'Manchester City'].Wage
group2 = df[df.Club == 'Juventus'].Wage
group3 = df[df.Club == 'Al Batin'].Wage
group4 = df[df.Club == 'AS Monaco'].Wage

df_new = pd.DataFrame([group1.mean(),group2.mean(),group3.mean(),group4.mean()], columns=['Wage'], index = ['Manchester City','Juventus','Al Batin','AS Monaco'])
df_new
# Выбрать любые 4 клуба + 2-ая колонка с названиями клубов.
```

```python
# показать по одному самому дорогому футболисту по убыванию в каждом клубе
df_new = pd.DataFrame(df,columns=['Club','Wage'] )
df_new
```

```python
# Среднее кол-во пенальти у игрока, возраст которого выше среднего, а его место на поле - ST.
df[(df.Position == 'ST') & (df.Age > df.Age.mean())].Penalties.mean()
```

```python
# Максимальная зарплата игрока, возраст котрого меньше среднего возраста, а скорость бега выше средней скорости бега, среди всех клубов.
df[(df.Age < df.Age.mean()) & (df.SprintSpeed > df.SprintSpeed.mean())].Wage.max()
```

```python
# Минимальная зарплата игрока клуба Ювентус, возраст которого выше среднего возраста и скорость бега тоже выше средней.
df[(df.Age > df.Age.mean()) & (df.Club == 'Juventus') & (df.SprintSpeed > df.SprintSpeed.mean())].Wage.min()
```

## Сводные таблицы

```python
df.loc[df['Club'].isin(['Real Madrid','FC Barcelona','Juventus','Manchester United'])]\
    .pivot_table(
        values=['Wage'],
        index=['Nationality'],
        columns=['Club'],
        aggfunc='sum'
    )
```
